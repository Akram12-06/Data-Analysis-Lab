```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_excel('C:\\Users\\exam2\\Downloads\\Adidas_US_Sales_Datasets_1_.xlsx')
dataset
```

1.	List all the retailers with retailer id.

```python
dataset.groupby(['Retailer','Retailer ID'])['Retailer'].apply(list)
```

2.	List all the retailers in every region.

```python
dataset.groupby('Retailer')['Region'].apply(list)
```

3.	List the retailers in every city of a state.

```python
dataset.groupby(['City','State'])['Retailer'].apply(list)
```

4.	List the products sold by the retailer in every city.

```python
dataset.groupby(['Retailer','City'])['Product'].apply(list)
```

5.	Find the total sales of every retailer.

```python
dataset.groupby('Retailer').sum('Total Sales')['Total Sales']
```

6.	Find the total sales of the retailers in every city along with profit.

```python
dataset.groupby(['Retailer','City']).sum(['Total Sales','Operating Margin'])[['Total Sales','Operating Profit']]
```

7.	Find the total sales &profit of each product sold by the retailer.

```python
dataset.groupby(['Product','Retailer']).sum(['Total Sales','Operating Profit'])[['Total Sales','Operating Profit']].apply(list)
# dataset
```

8.	Find the units sold, total sales &profit of the products sold between the dates 1/1/2020 and 4/15/2020.

```python
dataset[dataset['Invoice Date'].between('2020-01-01','2020-04-05')][['Units Sold','Total Sales','Operating Profit']]
```

9.	Find the no of units sold of each product by each retailer in every city.

```python
dataset.groupby(['Product','Retailer','City']).sum('Units Sold')['Units Sold']
```

10.	Find the products with different price per unit in different cities with proper information.

```python
dataset.groupby(['Product','City'])['Price per Unit'].first()
```

11.	Find the total sales &profits of all products in every month.

```python
dataset.groupby([pd.Grouper(key='Invoice Date',axis=0,freq='M'),'Product'])['Total Sales','Operating Profit'].sum()
```

```python
dataset.groupby([pd.Grouper(key='Invoice Date',axis=0,freq='M'),'Product'])['Total Sales','Operating Profit'].sum()
```

12.	Find the total sales &profit of the products in different sales methods in each city.

```python
dataset.groupby(['Sales Method','City']).sum(['Total Sales','Operating Profit'])[['Total Sales','Operating Profit']]
```

13.	Find the retailers who sold the same product with different prices in different cities.

```python
dataset.groupby(['Retailer','Product','City','Price per Unit'])['Retailer'].apply(list)
# dataset
```

14.	Find the products whose sales raises in every month.

```python
dataset.groupby(['Product',pd.Grouper(key='Invoice Date',axis=0,freq='M')])['Total Sales'].sum().diff().apply(lambda x: x>0)
```

15.	Find the retailers whose profit increased every month.

```python
dataset.groupby(['Retailer',pd.Grouper(key='Invoice Date',axis=0,freq='M')]).sum(['Operating Profit'])['Operating Profit'].diff().apply(lambda x:x>0)
```
