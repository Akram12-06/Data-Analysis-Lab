1.	 (i). Find out the no of movies released in every month of the year 1995.

```python
dataset
```

```python
movies=dataset[:]
```

```python
movies.drop(movies[movies['Release Date']=='TBD'].index,axis=0,inplace=True)
```

```python
movies['Release Date']=movies['Release Date'].apply(lambda x:pd.to_datetime(x))
```

```python
df=movies[movies['Release Date'].apply(lambda x:x.year)==1995]
df.groupby(movies['Release Date'].dt.month)['Title'].count()
```

(ii). Find out the no. of movies released in every year from 1990 to 1998.

```python
data=movies[movies['Release Date'].apply(lambda x:x.year).between(1990,1998)]
data.groupby(movies['Release Date'].dt.year)['Title'].count()
```
